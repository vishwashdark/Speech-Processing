```python
"""A1.Use librosa.effects.trim()to remove the silence parts of speech from beginning and end of your 
recorded signal.Listen to the new signal and perceptually compare the audio with original."""
import librosa
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
audio="03-01-01-01-01-01-01.wav"

y, sr = librosa.load(audio)
ipd.Audio(audio)

print(y.shape)
print(sr)

```

```python
y, sr = librosa.load(audio)
print("y: ", y[:10])
print(y.shape)
print(sr)
```

```python
import pandas as pd
a1 = pd.Series(y)
a1.plot(figsize=(10, 5),
       lw=1,
       title='Raw Audio')
plt.show()
```

```python
y_trimmed, _ = librosa.effects.trim(y, top_db=20)
pd.Series(y_trimmed).plot(figsize=(10,5),
                         lw=1,
                         title="Raw audio trimmed")
plt.show()
```
