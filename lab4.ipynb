```python
"""A3 Perform the spectral analysis of a word present in the recorded speech. 
Compare the spectrum with the spectrum of the full signal."""
import librosa
import numpy as np
import matplotlib.pyplot as plt
y, sr = librosa.load('Whatsapp Ptt 2024-02-01 at 1.42.46 PM.ogg')

start_sample = 5000
end_sample = 10000
word_signal = y[start_sample:end_sample]
print(word_signal.shape)
plt.plot(word_signal)
```

```python
word_spectrum = np.abs(librosa.stft(word_signal))
plt.plot(word_spectrum)
```

```python
#Specturm of one word of signal
plt.figure(figsize=(12, 6))
librosa.display.specshow(librosa.amplitude_to_db(word_spectrum, ref=np.max), y_axis='log', x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrum of Word Signal')
plt.tight_layout()
plt.show()
```

```python
#Specturn of one Full signal
full_spectrum = np.abs(librosa.stft(y))
plt.figure(figsize=(12, 6))
librosa.display.specshow(librosa.amplitude_to_db(full_spectrum, ref=np.max), y_axis='log', x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrum of Full Signal')
plt.tight_layout()
plt.show()
```

```python
"""A4. Take a rectangular window of 20 mili-second sampled at 22.5 KHz.
 Using FFT, analyse the spectral components."""

import numpy as np
import matplotlib.pyplot as plt
import librosa

fs = 22500 
T = 0.2 
t = np.linspace(0, T, int(fs*T))
full_spectrum = np.abs(librosa.stft(t))
plt.figure(figsize=(10, 5))
librosa.display.specshow(librosa.amplitude_to_db(full_spectrum, ref=np.max), x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrum of Full Signal')
plt.tight_layout()
plt.show()
```

```python
sample_rate = 22500  # 22.5 KHz
window_size = int(0.02 * sample_rate)  

windowed_signal = y[:window_size]
fft_result = np.fft.fft(windowed_signal)
amplitude_spectrum = np.abs(fft_result)
print(amplitude_spectrum)
frequencies = np.fft.fftfreq(window_size, 1/sample_rate)
print(frequencies)

# Plot the amplitude spectrum
plt.figure(figsize=(10, 4))
plt.plot(frequencies, amplitude_spectrum)
plt.title('Amplitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.show()
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
y, sr = librosa.load('Whatsapp Ptt 2024-02-01 at 1.42.46 PM.ogg')

window_length = int(sr * 0.02)  
hop_length = int(window_length / 2)  
frequency_matrix = []

# Iterate over windows
for i in range(0, len(y) - window_length, hop_length):
    window = y[i:i + window_length]
    D = librosa.stft(window)

    # Append the magnitude spectrum to the frequency matrix
    frequency_matrix.append(np.abs(D))
```

```python
frequency_matrix = np.array(frequency_matrix)

# Plot the frequency matrix as a heatmap
plt.imshow(frequency_matrix, cmap='viridis', aspect='auto', origin='lower', extent=[0, len(y), 0, sr/2])
plt.colorbar(format='%+2.0f dB')
plt.title('Frequency Components Heatmap')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.show()
```
