```python
"""A1. Find the first derivative of your speech signal with finite difference method. Listen to the first
derivative signal and the original speech signal."""
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

y, sr = librosa.load("WhatsApp Ptt 2024-02-01 at 1.42.46 PM.ogg")
dy = np.zeros_like(y)

for i in range(1, len(y)):
    dy[i-1] = y[i] - y[i-1]

plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
librosa.display.waveshow(y, sr=sr)
plt.title('Original Speech Signal')

plt.subplot(2, 1, 2)
librosa.display.waveshow(dy, sr=sr)
plt.title('First Derivative of Speech Signal')
plt.tight_layout()
plt.show()
```

```python
"""A2. Detect the points of zero crossing in the first derivative signal. Compare the average length
between two consecutive zero crossings for speech and silence regions. Observe the pattern."""

import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

y, sr = librosa.load("WhatsApp Ptt 2024-02-01 at 1.42.46 PM.ogg")
dy = np.diff(y, n=1)  # First derivative
dpy = np.diff(y, n=2)  # Second derivative

dpy = np.concatenate((dpy, [0]))
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
librosa.display.waveshow(y, sr=sr)
plt.title('Original Speech Signal')

plt.subplot(3, 1, 2)
librosa.display.waveshow(dy, sr=sr)
plt.title('First Derivative of Speech Signal')
plt.tight_layout()

plt.subplot(3, 1, 3)
librosa.display.waveshow(dpy, sr=sr,color="red")
plt.title('Second Derivative of Speech Signal')
plt.tight_layout()

plt.show()
```

```python
D = librosa.stft(y_stat)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
S_db.shape
# Plot the transformed audio data
fig, ax = plt.subplots(figsize=(10, 5))
img = librosa.display.specshow(S_db,
                              x_axis='time',
                              y_axis='log',
                              ax=ax)
ax.set_title('Spectogram for Statment', fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
plt.show()
```
